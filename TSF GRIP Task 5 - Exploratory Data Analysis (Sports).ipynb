### Number of matches won by winning team each season

```python
teams_season = dataset.groupby('season')['winner'].value_counts()
year = 2008
win_season_df = pd.DataFrame(columns=['year', 'Team', 'Number of Wins'])
for items in teams_season.iteritems():    
    if items[0][0]==year:
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'Team': [items[0][1]],
            'Number of Wins': [items[1]]
        })
        win_season_df = win_season_df.append(win_series)
        year += 1

plt.rcParams.update({'font.size': 20})
fig = plt.figure(facecolor='#07000d')
ax1 = fig.add_subplot(1, 1, 1, facecolor='#07000d')
plt.rcParams['text.color'] = '#FFFFFF'
plt.xlabel('Team 2', color='#ffffff')
plt.ylabel('Count', color='#ffffff')
ax1.tick_params(axis='y', colors='#ffffff')
ax1.tick_params(axis='x', colors='#ffffff')
plt.title("Number of matches won by winning team each season")

sns.barplot('Team','Number of Wins',data=win_season_df,hue='year',palette='dark')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Venue which hosted the maximum number of Matches

```python
venue_mat=dataset['venue'].value_counts()
venue_data = pd.DataFrame(columns=['Venue', 'Number of Matches'])
for items in venue_mat.iteritems():
    temp_df = pd.DataFrame({
        'Venue':[items[0]],
        'Number of Matches':[items[1]]
    })
    venue_data = venue_data.append(temp_df, ignore_index=True)
    
plt.rcParams.update({'font.size': 20})
fig = plt.figure(facecolor='#07000d')
ax1 = fig.add_subplot(1, 1, 1, facecolor='#07000d')
plt.rcParams['text.color'] = '#FFFFFF'
plt.xlabel('Team 2', color='#ffffff')
plt.ylabel('Count', color='#ffffff')
ax1.tick_params(axis='y', colors='#ffffff')
ax1.tick_params(axis='x', colors='#ffffff')
plt.title("Venue which hosted the maximum number of Matches")


sns.barplot(x='Venue',y='Number of Matches',data=venue_data[:5],palette='dark')
plt.xticks(rotation=45,fontsize=16, ha='right')
plt.show()
```

### Team which won Toss the most

```python
toss_ser = dataset['toss_winner'].value_counts()

toss_df = pd.DataFrame(columns=["team", "wins"])

for items in toss_ser.iteritems():
    temp_df3 = pd.DataFrame({
        'team':[items[0]],
        'wins':[items[1]]
    })
    toss_df = toss_df.append(temp_df3, ignore_index=True)
    
plt.rcParams.update({'font.size': 20})
fig = plt.figure(facecolor='#07000d')
ax1 = fig.add_subplot(1, 1, 1, facecolor='#07000d')
plt.rcParams['text.color'] = '#FFFFFF'
plt.xlabel('Team 2', color='#ffffff')
plt.ylabel('Count', color='#ffffff')
ax1.tick_params(axis='y', colors='#ffffff')
ax1.tick_params(axis='x', colors='#ffffff')
plt.title("Team which won Toss the most")

sns.barplot(x='wins', y='team', data=toss_df, palette='Paired');
plt.show()
```

### Player of the Match 

```python
plt.rcParams.update({'font.size': 20})
fig = plt.figure(facecolor='#07000d')
ax1 = fig.add_subplot(1, 1, 1, facecolor='#07000d')
plt.rcParams['text.color'] = '#FFFFFF'
plt.xlabel('Team 2', color='#ffffff')
plt.ylabel('Count', color='#ffffff')
ax1.tick_params(axis='y', colors='#ffffff')
ax1.tick_params(axis='x', colors='#ffffff')
plt.title("Player of the Match")

fav_cities = dataset['player_of_match'].value_counts().reset_index()
fav_cities.columns = ['player_of_match','count']
sns.barplot(x = 'player_of_match',y = 'count', data = fav_cities[:20])
plt.xticks(rotation=45, ha='right')
plt.show()
```
